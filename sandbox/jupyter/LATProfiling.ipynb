# LAT Profiling output

We profile the two different versions of LAT using cProfile and saved all functions calls with a cumulative time greater than 1 second into a dataframe.

```python
import pandas as pd
from IPython.display import Image

pd.options.display.max_rows = 100
```

# LAT-v1 Profile breakdown
Total processing time is 191.54 minutes for 5000 events, or 26.1 events/minute

The processing is dominated at 98% by the waveform fitting

```python
Image(filename='LATv1_Profile.png') 
```

# LAT-v2 Profile breakdown
Total processing time is 19.95 minutes for 5000 events, or 250.7 events/minute

The processing is dominated at 80.7% by the waveform fitting, 6.6% by the asymmetric trapezoid, and 5.2% by the regular trapezoid.

The asymmetric trapezoid is about 3 times slower than the standard trapezoid as it doesn't use a recursive algorithm.

```python
Image(filename='LATv2_Profile.png') 
```

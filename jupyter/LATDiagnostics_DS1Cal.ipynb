```python
# Diagnostic plots for DS1 calibration files
import ROOT
import numpy as np
ROOT.gROOT.ProcessLine("gErrorIgnoreLevel = 3001;")
ROOT.gStyle.SetOptStat(0)
```

```python
# Draw two canvases per cut combination, left is mHL == 1, right is mHL == 2
# Highlight one channel (ch 626) in blue
theCut = "trapENFCal>1 && (mHL==1 || mHL==2) && gain==0 && isGood && !muVeto && !(C==1&&isLNFill1) && !(C==2&&isLNFill2) && C!=0&&P!=0&&D!=0"

# Load runs 10241 to 10258
skimTree = ROOT.TChain("skimTree")
skimTree.Add("~/project/cal-lat/latSkimDS1_run*.root")
skimTree.GetEntries(theCut)
```

```python
# bcMax
nP = 2
c1 = ROOT.TCanvas("c1","c1",nP*700,600); c1.Divide(nP,1,0)
xBins, xBinL, xBinH = 100, 0, 25
yBins, yBinL, yBinH = 10000, 100, 1000
xStr, yStr = 'trapENFCal', 'bcMax'
h1=ROOT.TH2D("h1","",xBins,xBinL,xBinH,yBins,yBinL,yBinH)
skimTree.Project("h1",yStr+":"+xStr,theCut+"&& mHL==1")
h2=ROOT.TH2D("h2","",xBins,xBinL,xBinH,yBins,yBinL,yBinH)
skimTree.Project("h2",yStr+":"+xStr,theCut+"&& mHL==2")
h1.GetXaxis().SetTitle(xStr); h1.GetYaxis().SetTitle(yStr)
h2.GetXaxis().SetTitle(xStr)
h1.SetMarkerColorAlpha(ROOT.kBlack, 0.5)
h2.SetMarkerColorAlpha(ROOT.kBlack, 0.5)
c1.cd(1);ROOT.gPad.SetLogy(0);h1.Draw()
c1.cd(2);ROOT.gPad.SetLogy(0);h2.Draw()
h3=ROOT.TH2D("h3","",xBins,xBinL,xBinH,yBins,yBinL,yBinH)
skimTree.Project("h3",yStr+":"+xStr,theCut+"&& mHL==1 && channel == 626")
h4=ROOT.TH2D("h4","",xBins,xBinL,xBinH,yBins,yBinL,yBinH)
skimTree.Project("h4",yStr+":"+xStr,theCut+"&& mHL==2 && channel == 626")
c1.cd(1); h3.SetMarkerColorAlpha(ROOT.kBlue, 0.35); h3.SetMarkerStyle(7); h3.Draw("SAME")
c1.cd(2); h4.SetMarkerColorAlpha(ROOT.kBlue, 0.35); h4.SetMarkerStyle(7); h4.Draw("SAME")
c1.Draw()
```
